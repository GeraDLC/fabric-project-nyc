# Obtener datos hist√≥ricos desde la web (desde 2024)

```python
from datetime import datetime
from dateutil.relativedelta import relativedelta
import requests
import os

# Par√°metros
tipo = "yellow"
inicio = datetime(2024, 6, 1)
fin = datetime.today().replace(day=1)

# Rango de fechas
def generar_rango_fechas(inicio, fin):
    fechas = []
    actual = inicio
    while actual <= fin:
        fechas.append(actual)
        actual += relativedelta(months=1)
    return fechas
```

Se verifica que el archivo exista, de lo contrario se descarga.

```python
# Iterar sobre los meses
for fecha in generar_rango_fechas(inicio, fin):
    anio_mes = fecha.strftime("%Y-%m")
    nombre_archivo = f"{tipo}_tripdata_{anio_mes}.parquet"
    ruta_lakehouse = f"/lakehouse/default/Files/nyctaxi_yellow/{nombre_archivo}"
    url = f"https://d37ci6vzurychx.cloudfront.net/trip-data/{nombre_archivo}"

    if not os.path.exists(ruta_lakehouse):
        print(f"[{anio_mes}] Descargando {nombre_archivo}...\n")

        response = requests.get(url)
        if response.status_code == 200:
            with open(ruta_lakehouse, "wb") as f:
                f.write(response.content)
            print(f"‚úÖ {nombre_archivo} descargado.\n")
        else:
            print(f"‚ö†Ô∏è No se pudo descargar {nombre_archivo} - C√≥digo HTTP: {response.status_code}\n Probablemente a√∫n no exista.\n")
    else:
        print(f"üü° {nombre_archivo} ya existe. Se omite.\n")
```
